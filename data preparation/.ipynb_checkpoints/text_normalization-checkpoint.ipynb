```python
import numpy as np
import pand
as as pd
import glob
import nltk
import re
from tqdm import tqdm
```

### Load mtsamples data

```python
mtsamples_path =  "data/mtsamples.csv"
df_mtsamples = pd.read_csv(mtsamples_path ,index_col=False)
df_mtsamples = df_mtsamples.dropna(subset = ['transcription']).reset_index(drop=True).drop(columns=['Unnamed: 0'])
df_mtsamples.head()
```

```python
df_mtsamples.transcription[0]
```

```python
# Add transcriptions to list
sentences = []
for i in range(len(df_mtsamples.transcription)):
    text = df_mtsamples.transcription[i]
    sentences.extend(nltk.tokenize.sent_tokenize(text))
```

### Make text normalization

#### Normalization. Step 1

```python
new_sentences=[]
for i in tqdm(range(len(sentences))):
    sentences[i]=sentences[i].replace('.,', '. ')
    sentences[i]=sentences[i].replace(':,', ': ')
    sentences[i]=sentences[i].replace(':,', ': ')
    sentences[i]=sentences[i].replace(': , ', ': ')
    sentences[i]=sentences[i].replace(': ,', ': ')
    sentences[i]=sentences[i].replace('  ', ' ')
    
    sentences_batch = nltk.tokenize.sent_tokenize(sentences[i])
    for sentence in sentences_batch:
        if len((sentence.replace('.', '').replace(':', '').replace(',','')).split(' '))>2:
            new_sentences.append(sentence)
```

####  Normalization. Step 2

```python
import re
import inflect
from unidecode import unidecode

_inflect = inflect.engine()
_comma_number_re = re.compile(r"([0-9][0-9\,]+[0-9])")
_decimal_number_re = re.compile(r"([0-9]+\.[0-9]+)")
_pounds_re = re.compile(r"£([0-9\,]*[0-9]+)")
_dollars_re = re.compile(r"\$([0-9\.\,]*[0-9]+)")
_ordinal_re = re.compile(r"[0-9]+(st|nd|rd|th)")
_number_re = re.compile(r"[0-9]+")

_number_mm_re = re.compile(r"([0-9\,]*[0-9])mm")
_number_cm_re = re.compile(r"([0-9\,]*[0-9])cm")
_number_mg_re = re.compile(r"([0-9\,]*[0-9])mg")
_number_kg_re = re.compile(r"([0-9\,]*[0-9])kg")
_number_ml_re = re.compile(r"([0-9\,]*[0-9])ml")
_number_mL_re = re.compile(r"([0-9\,]*[0-9])mL")
_number_dl_re = re.compile(r"([0-9\,]*[0-9])dl")
_number_dL_re = re.compile(r"([0-9\,]*[0-9])dl")
_number_ng_re = re.compile(r"([0-9\,]*[0-9])ng")
_number_ug_re = re.compile(r"([0-9\,]*[0-9])ug")

_chars_number_re = re.compile("([a-zA-Z]+)([0-9]+)")

_chars_punct_re = re.compile("([\.\,\:\;\)]+)([a-zA-Z]+)")

def _remove_commas(m):
    return m.group(1).replace(",", "")


def _expand_decimal_point(m):
    return m.group(1).replace(".", " point ")


def _expand_dollars(m):
    match = m.group(1)
    parts = match.split(".")
    if len(parts) > 2:
        return match + " dollars"  # Unexpected format
    dollars = int(parts[0]) if parts[0] else 0
    cents = int(parts[1]) if len(parts) > 1 and parts[1] else 0
    if dollars and cents:
        dollar_unit = "dollar" if dollars == 1 else "dollars"
        cent_unit = "cent" if cents == 1 else "cents"
        return "%s %s, %s %s" % (dollars, dollar_unit, cents, cent_unit)
    elif dollars:
        dollar_unit = "dollar" if dollars == 1 else "dollars"
        return "%s %s" % (dollars, dollar_unit)
    elif cents:
        cent_unit = "cent" if cents == 1 else "cents"
        return "%s %s" % (cents, cent_unit)
    else:
        return "zero dollars"


def _expand_ordinal(m):
    return _inflect.number_to_words(m.group(0))


def _expand_number(m):
    num = int(m.group(0))
    if num > 1000 and num < 3000:
        if num == 2000:
            return "two thousand"
        elif num > 2000 and num < 2010:
            return "two thousand " + _inflect.number_to_words(num % 100)
        elif num % 100 == 0:
            return _inflect.number_to_words(num // 100) + " hundred"
        else:
            return _inflect.number_to_words(num, andword="", zero="oh", group=2).replace(", ", " ")
    else:
        return _inflect.number_to_words(num, andword="")


def normalize_numbers(text):
    text = re.sub(_comma_number_re, _remove_commas, text)
    text = re.sub(_pounds_re, r"\1 pounds", text)
    
    text = re.sub(_number_mm_re, r"\1 millimetres", text)
    text = re.sub(_number_cm_re, r"\1 centimeters", text)
    text = re.sub(_number_mg_re, r"\1 milligrams", text)
    text = re.sub(_number_kg_re, r"\1 kilograms", text)
    text = re.sub(_number_ml_re, r"\1 milliliters", text)
    text = re.sub(_number_mL_re, r"\1 milliliters", text)
    text = re.sub(_number_dl_re, r"\1 deciliters", text)
    text = re.sub(_number_dL_re, r"\1 deciliters", text)
    text = re.sub(_number_ng_re, r"\1 nanograms", text)
    text = re.sub(_number_ug_re, r"\1 micrograms", text)
    
    text = re.sub(_chars_number_re, r"\1 \2", text)
    
    text = re.sub(_dollars_re, _expand_dollars, text)
    text = re.sub(_decimal_number_re, _expand_decimal_point, text)
    text = re.sub(_ordinal_re, _expand_ordinal, text)
    text = re.sub(_number_re, _expand_number, text)
    
    text = re.sub(_chars_punct_re, r"\1 \2", text)
    
    return text

# Regular expression matching whitespace:
_whitespace_re = re.compile(r"\s+")

# List of (regular expression, replacement) pairs for abbreviations:
_abbreviations = [(re.compile("\\b%s\\." % x[0], re.IGNORECASE), x[1]) for x in [
  ("mrs", "misess"),
  ("mr", "mister"),
  ("dr", "doctor"),
  ("st", "saint"),
  ("co", "company"),
  ("jr", "junior"),
  ("maj", "major"),
  ("gen", "general"),
  ("drs", "doctors"),
  ("rev", "reverend"),
  ("lt", "lieutenant"),
  ("hon", "honorable"),
  ("sgt", "sergeant"),
  ("capt", "captain"),
  ("esq", "esquire"),
  ("ltd", "limited"),
  ("col", "colonel"),
  ("ft", "fort"),
    
  ("cm", "centimeters"),
  ("mm", "millimetres"),
  ("p.m.", "after noon"),
  ("a.m.", "before noon"),
  ("mmHg", "millimetres of mercury"),
  ("mL", "milliliters"),
  ("dL", "deciliters"),
  ("ml", "milliliters"),
  ("cc", "cubic centimeter"),
  ("p.o", "per os"),
  ("p.o.s", "points of service"),
  ("kg", "kilograms"),
  ("Hz", "Hertz"),
  ("b.i.d.", "twice a day")

]]


def expand_abbreviations(text):
    for regex, replacement in _abbreviations:
        text = re.sub(regex, replacement, text)
    return text


def expand_numbers(text):
    return normalize_numbers(text)


def lowercase(text):
    """lowercase input tokens."""
    return text.lower()


def collapse_whitespace(text):
    return re.sub(_whitespace_re, " ", text)


def convert_to_ascii(text):
    return unidecode(text)


def basic_cleaners(text):
    """Basic pipeline that lowercases and collapses whitespace without transliteration."""
    #text = lowercase(text)
    text = collapse_whitespace(text)
    return text


def transliteration_cleaners(text):
    """Pipeline for non-English text that transliterates to ASCII."""
    text = convert_to_ascii(text)
    #text = lowercase(text)
    text = collapse_whitespace(text)
    return text


def english_cleaners(text):
    """Pipeline for English text, including number and abbreviation expansion."""
    text = convert_to_ascii(text)
    #text = lowercase(text)
    text = expand_numbers(text)
    text = expand_abbreviations(text)
    text = collapse_whitespace(text)
    return text
```

```python
normolized_sentences=[]
for i in tqdm(range(len(new_sentences))):
    
    normalized_sentence = english_cleaners(new_sentences[i])
    normalized_sentence = normalized_sentence.replace('#', 'number ')
    normalized_sentence = normalized_sentence.replace('one%', 'one percent')
    normalized_sentence = normalized_sentence.replace('%', ' percents')
    
    normalized_sentence = normalized_sentence.replace('doctor XYZ', 'dortor Artem')
    normalized_sentence = normalized_sentence.replace('doctor XXX', 'dortor Artem')
    normalized_sentence = normalized_sentence.replace('doctor X', 'dortor Artem')
    normalized_sentence = normalized_sentence.replace('doctor Y', 'dortor Artem')
    normalized_sentence = normalized_sentence.replace('doctor Z', 'dortor Ivan')
    normalized_sentence = normalized_sentence.replace('doctor W', 'dortor Vlad')
    normalized_sentence = normalized_sentence.replace('M.D.', 'dortor of medicine')
    normalized_sentence = normalized_sentence.replace('-mm', '-millimetres')
    normalized_sentence = normalized_sentence.replace('Month DD, YYYY', 'first of september twenty seven')
    normalized_sentence = normalized_sentence.replace(' x ', ' on ')
    normalized_sentence = normalized_sentence.replace('_', '')
    normalized_sentence = normalized_sentence.replace("Mrs", "misess")
    normalized_sentence = normalized_sentence.replace("Mr", "mister")
    normalized_sentence = normalized_sentence.replace("Ms", "miss")
    
    
    normalized_sentence = normalized_sentence.replace(" cm ", " centimeters ")
    normalized_sentence = normalized_sentence.replace(" mm ", " millimetres ")
    normalized_sentence = normalized_sentence.replace("p.m.", "after noon")
    normalized_sentence = normalized_sentence.replace("a.m.", "before noon")
    normalized_sentence = normalized_sentence.replace("mmHg", "millimetres of mercury")
    normalized_sentence = normalized_sentence.replace(" mL ", " milliliters ")
    normalized_sentence = normalized_sentence.replace(" Hz ", " hertz ")
    normalized_sentence = normalized_sentence.replace(" ml ", " milliliters ")
    normalized_sentence = normalized_sentence.replace("dL", "deciliters")
    normalized_sentence = normalized_sentence.replace(" dL ", " deciliters ")
    normalized_sentence = normalized_sentence.replace(" DL ", " deciliters ")
    normalized_sentence = normalized_sentence.replace(" cc ", " cubic centimeter ")
    normalized_sentence = normalized_sentence.replace("p.o", "per os")
    normalized_sentence = normalized_sentence.replace("p.o.s", "points of service")
    normalized_sentence = normalized_sentence.replace("q.d", "once a day")
    normalized_sentence = normalized_sentence.replace(" kg ", " kilograms ")
    normalized_sentence = normalized_sentence.replace(" mg ", " milligrams ")
    normalized_sentence = normalized_sentence.replace("b.i.d.", "twice a day") 
    normalized_sentence = normalized_sentence.replace(" pH", " Hydrogen Ion Concentration")
    normalized_sentence = normalized_sentence.replace("/", " ")
    normalized_sentence = normalized_sentence.replace("\'", " ")
    normalized_sentence = normalized_sentence.replace(': ,', ": ")
    normalized_sentence = normalized_sentence.replace('XXXX', "")
    normalized_sentence = normalized_sentence.replace('XXX', "")
    normalized_sentence = normalized_sentence.replace('XX', "twenty-six")
    normalized_sentence = normalized_sentence.replace('<', " less than ")
    normalized_sentence = normalized_sentence.replace('>', " greater than ")
    normalized_sentence = normalized_sentence.replace('=', " equal to ")
    normalized_sentence = normalized_sentence.replace('(+)', "")
    normalized_sentence = normalized_sentence.replace('(-)', "")
    normalized_sentence = normalized_sentence.replace('+', " plus")
    
    normalized_sentence=normalized_sentence.strip()
    
    if normalized_sentence[0]==',':
        normalized_sentence=normalized_sentence[1:]
        
    normolized_sentences.append(normalized_sentence)
```

```python
def get_all_caps_words(text):
    """ Get all words with capitalixation from text.
        inputs: text - string of text.
        returns: list of words with capitalization.
    """
    return re.findall(r"(\b(?:[A-Z]+[a-z]?[A-Z]*|[A-Z]*[a-z]?[A-Z]+)\b(?:\s+(?:[A-Z]+[a-z]?[A-Z]*|[A-Z]*[a-z]?[A-Z]+)\b)*)",text)
```

```python
abbs_list = []

for i in range(len(normolized_sentences)):
    text = normolized_sentences[i]
    abbs_for_text = get_all_caps_words(text)
    for abb in abbs_for_text:
        if len(abb)<5 and len(abb)>1:
            abbs_list.append(abb)
            
adds_dict=dict(zip(abbs_list,[abbs_list.count(i) for i in abbs_list]))
```

```python
len_threshold=110 
part_threshold=30

def text2fragments(text, len_threshold=len_threshold):
    """
        Sorry for unreadable code. This function requires refactoring!
        Split text on sentences and big sentences (with punctuation only) on smaller sentences.
        Input: text - string of text
        Returns: fragments - list of sentences, 
                 fragments_punctuation_end - list of the last punctuation sign for each fragment
    """
    
    fragments = []
    fragments_punctuation_end = []
    punctuation_types = ['tab', 'point', 'colon', 'semicolon', 'comma']
    text=text.replace('\n\n', '\n')
    paragraphs = text.split('\n')
    
    for i in (range(len(paragraphs))):
        paragraph = paragraphs[i]
        splited_paragraph = nltk.tokenize.sent_tokenize(paragraph)
        
        for j in range(len(splited_paragraph)):
            sentence = splited_paragraph[j]
            
            splited_sentence_by_colon = sentence.split(':')
            
            if len(splited_sentence_by_colon)>1:
                for k1 in range(len(splited_sentence_by_colon)):
                    part_1 = splited_sentence_by_colon[k1]
                    
                    splited_part_by_semicolon = part_1.split(';')
                    if len(splited_part_by_semicolon)>1:
                        for k2 in range(len(splited_part_by_semicolon)):
                            
                            part_2 = splited_part_by_semicolon[k2]
                            if part_2!='':
                                if len(part_2)<len_threshold:
                                    fragments.append(part_2)
                                    if k2==0:
                                        fragments_punctuation_end.append(punctuation_types[2])# add colon
                                    elif k2<len(splited_sentence_by_colon)-1 and k2!=0:
                                        fragments_punctuation_end.append(punctuation_types[3])# add semicolon
                                    else:
                                        if j==len(splited_paragraph)-1:
                                            fragments_punctuation_end.append(punctuation_types[0])# add tab
                                        else:
                                            fragments_punctuation_end.append(punctuation_types[1])# add point
                                elif sentence_spliter(part_2)==None:
                                    fragments.append(part_2)
                                    if j==len(splited_paragraph)-1:
                                        fragments_punctuation_end.append(punctuation_types[0])# add tab
                                    else:
                                        fragments_punctuation_end.append(punctuation_types[1])# add point
                                else:
                                    splited_sentence = sentence_spliter(part_2)
                                    for s in range(len(splited_sentence)):
                                        if s< len(splited_sentence)-1:
                                            fragments.append((splited_sentence[s] + ', ').strip())
                                            fragments_punctuation_end.append(punctuation_types[4])#add comma
                                        else:
                                            fragments.append(splited_sentence[s].strip())
                                            if j==len(splited_paragraph)-1:
                                                fragments_punctuation_end.append(punctuation_types[0])# add tab
                                            else:
                                                fragments_punctuation_end.append(punctuation_types[1])# add point
                    else:
                        # we don't have semicolon
                        if part_1!='':
                            if len(part_1)<len_threshold:
                                fragments.append(part_1)
                                if k1<len(splited_sentence_by_colon)-1:
                                    fragments_punctuation_end.append(punctuation_types[2])# add colon
                                else:
                                    if j==len(splited_paragraph)-1:
                                        fragments_punctuation_end.append(punctuation_types[0])# add tab
                                    else:
                                        fragments_punctuation_end.append(punctuation_types[1])# add point
                            elif sentence_spliter(part_1)==None:
                                fragments.append(part_1)
                                if j==len(splited_paragraph)-1:
                                    fragments_punctuation_end.append(punctuation_types[0])# add tab
                                else:
                                    fragments_punctuation_end.append(punctuation_types[1])# add point
                            else:
                                splited_sentence = sentence_spliter(part_1)
                                for s in range(len(splited_sentence)):
                                    if s< len(splited_sentence)-1:
                                        fragments.append((splited_sentence[s] + ', ').strip())
                                        fragments_punctuation_end.append(punctuation_types[4])#add comma
                                    else:
                                        fragments.append(splited_sentence[s].strip())
                                        if j==len(splited_paragraph)-1:
                                            fragments_punctuation_end.append(punctuation_types[0])# add tab
                                        else:
                                            fragments_punctuation_end.append(punctuation_types[1])# add point
                                        
            else:
                # we don't have colon but have semicolon
                splited_sentence_by_semicolon = sentence.split(';')
                if len(splited_sentence_by_semicolon)>1:
                    
                    for k3 in range(len(splited_sentence_by_semicolon)):
                        
                        part_3 = splited_sentence_by_semicolon[k3]
                        if part_3!='':
                            if len(part_3)<len_threshold:
                                fragments.append(part_3)
                                if k3<len(splited_sentence_by_semicolon)-1:
                                    fragments_punctuation_end.append(punctuation_types[3])# add semicolon
                                else:
                                    if j==len(splited_paragraph)-1:
                                        fragments_punctuation_end.append(punctuation_types[0])# add tab
                                    else:
                                        fragments_punctuation_end.append(punctuation_types[1])# add point
                            elif sentence_spliter(part_3)==None:
                                fragments.append(part_3)
                                if j==len(splited_paragraph)-1:
                                    fragments_punctuation_end.append(punctuation_types[0])# add tab
                                else:
                                    fragments_punctuation_end.append(punctuation_types[1])# add point
                            else:
                                splited_sentence = sentence_spliter(part_3)
                                for s in range(len(splited_sentence)):
                                    if s< len(splited_sentence)-1:
                                        fragments.append((splited_sentence[s] + ', ').strip())
                                        fragments_punctuation_end.append(punctuation_types[4])# add comma
                                    else:
                                        fragments.append(splited_sentence[s].strip())
                                        if j==len(splited_paragraph)-1:
                                            fragments_punctuation_end.append(punctuation_types[0])# add tab
                                        else:
                                            fragments_punctuation_end.append(punctuation_types[1])# add point
                else:
                    # we don't have colon and semicolons
                    if sentence!='':
                        if len(sentence)<len_threshold:
                            fragments.append(sentence)
                            if j==len(splited_paragraph)-1:
                                fragments_punctuation_end.append(punctuation_types[0])# add tab
                            else:
                                fragments_punctuation_end.append(punctuation_types[1])# add point
                        elif sentence_spliter(sentence)==None:
                            fragments.append(sentence)
                            if j==len(splited_paragraph)-1:
                                fragments_punctuation_end.append(punctuation_types[0])# add tab
                            else:
                                fragments_punctuation_end.append(punctuation_types[1])# add point
                        else:
                            splited_sentence = sentence_spliter(sentence)
                            for s in range(len(splited_sentence)):
                                if s< len(splited_sentence)-1:
                                    fragments.append((splited_sentence[s] + ', ').strip())
                                    fragments_punctuation_end.append(punctuation_types[4])# add comma
                                else:
                                    fragments.append(splited_sentence[s].strip())
                                    if j==len(splited_paragraph)-1:
                                        fragments_punctuation_end.append(punctuation_types[0])# add tab
                                    else:
                                        fragments_punctuation_end.append(punctuation_types[1])# add point
        
    return fragments, fragments_punctuation_end


def sentence_spliter(sentence, len_threshold=len_threshold, part_threshold=part_threshold):  
    """
        Split big sentences (with punctuation only) on smaller sentences.
        Input: sentence - string of text
        Returns: sentence_part_list - list of sentences
    """
    if len(sentence)<len_threshold:
        return(sentence)
    else:
        splited_sent = sentence.split(',')
        if len(splited_sent)>1:
            sentence_parts = []
            sentence_part_list = []
            char_sum = 0
            sentence_part = ''
            counter = 0
            
            for i in range(len(splited_sent)):                
                if sentence_part=='':
                    sentence_part = sentence_part + splited_sent[i]    
                else:
                    sentence_part = sentence_part + ', '+ splited_sent[i]
                    counter += 1                   
                char_sum += len(sentence_part)
                
                if char_sum>part_threshold:
                    sentence_part_list.append(sentence_part)
                    sentence_part = ''
                    char_sum = 0
                elif char_sum<=part_threshold and i==len(splited_sent)-1:
                    sentence_part_list[-1]=sentence_part_list[-1]+', '+sentence_part                
                    
            return sentence_part_list 
```

####  Normalization. Step 5: Remove all long fragments

```python
# The sentences longer that 200 characters can't be read by TTS with local-sensitive attention.

normolized_sentences_3=[]
trash=[]

for i in tqdm(range(len(normolized_sentences_2))):
    text = normolized_sentences_2[i]
    if len(text)>200:
        trash.append(text)
    elif len(text)<10:
        trash.append(text)
    else:
        text = lowercase(text)
        text = expand_abbreviations(text)
        text = collapse_whitespace(text)
        text=text.strip()
        if text[0] in [',', '.', '"', ':', ';']:
            text=text[1:]
        text=text.strip()
        normolized_sentences_3.append(text)
```
